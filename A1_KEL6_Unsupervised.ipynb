#Business Understanding

Dataset "Countries of the World" dari Kaggle adalah kumpulan data yang berisi informasi mengenai berbagai negara di seluruh dunia. Dataset ini memiliki 20 kolom yang memberikan berbagai informasi.

Tujuan dari dataset ini adalah untuk membantu para analis dalam mempelajari dan memahami perbedaan antara negara-negara di seluruh dunia dan faktor-faktor yang memengaruhi kesejahteraan penduduk di negara-negara tersebut

**Penjelasan Kolom**

1. Country: Nama negara.

2. Region: Wilayah geografis di mana negara tersebut berada.

3. Population: Jumlah penduduk negara.

4. Area (sq. mi.): Luas wilayah negara dalam mil persegi.

5. Pop. Density (per sq. mi.): Kepadatan penduduk dalam mil persegi.

6. Coastline (coast/area ratio): Rasio garis pantai terhadap luas wilayah negara.

7. Net migration: Angka migrasi neto (jumlah orang yang pindah masuk dikurangi jumlah orang yang pindah keluar) dalam negara tersebut.

8. Infant mortality (per 1000 births): Angka kematian bayi per 1000 kelahiran hidup.

9. GDP ($ per capita): Produk domestik bruto (PDB) per kapita dalam dolar AS.

10. Literacy (%): Persentase populasi yang bisa membaca dan menulis.

11. Phones (per 1000): Jumlah telepon seluler per 1000 orang.

12. Arable (%): Persentase lahan yang dapat digunakan untuk pertanian.

13. Crops (%): Persentase lahan yang ditanami dengan tanaman.

14. Other (%): Persentase lahan yang tidak dapat digunakan untuk pertanian.

15. Climate: Klasifikasi iklim berdasarkan skala K√∂ppen, dengan nilai-nilai yang meliputi: 1 (tropis), 2 (subtropis), 3 (sedang), 4 (dingin), 5 (tundra), dan 6 (gurun).

16. Birthrate: Jumlah kelahiran per 1000 penduduk dalam setahun.

17. Deathrate: Jumlah kematian per 1000 penduduk dalam setahun.

18. Agriculture: Persentase PDB yang berasal dari sektor pertanian.

19. Industry: Persentase PDB yang berasal dari sektor industri.

20. Service: Persentase PDB yang berasal dari sektor jasa.

Dataset ini awalnya dikumpulkan dari berbagai sumber, termasuk CIA World Factbook, United Nations, dan World Health Organization, dan dapat digunakan untuk berbagai macam tujuan, seperti analisis data, visualisasi, atau pembuatan model prediksi.

###Menghilangkan outliers

Hal ini dilakukan untuk memastikan bahwa data yang digunakan dalam analisis atau model machine learning adalah data yang lebih representatif dan tidak terpengaruh oleh nilai-nilai ekstrem. Dengan menghilangkan outliers, dapat membantu meminimalkan kesalahan dalam analisis dan meningkatkan kinerja model machine learning yang dibangun.

```python
# Menghitung nilai IQR pada kolom 'GDP ($ per capita)'
Q1 = df1['GDP ($ per capita)'].quantile(0.25)
Q3 = df1['GDP ($ per capita)'].quantile(0.75)
IQR = Q3 - Q1

# Menghitung nilai batas bawah dan batas atas
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

# Menghapus baris dengan nilai di luar batas bawah dan batas atas
df1 = df1[(df1['GDP ($ per capita)'] >= lower_bound) & (df1['GDP ($ per capita)'] <= upper_bound)]
df1
```

#Modelling

###K-Means

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

def kmeans_clustering(data, n_clusters):
    # Preprocessing data
    data = data[['Country', 'Population', 'Area (sq. mi.)']].dropna()
    data['Density'] = data['Population'] / data['Area (sq. mi.)']
    X = data[['Density']]

    # Create k-means model
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=0)

    # Fit the model to the data
    kmeans.fit(X)

    # Assign the cluster labels to the original dataset
    data['Cluster'] = kmeans.labels_

    # Print the countries in each cluster
    for i in range(n_clusters):
        print(f'Cluster {i}:')
        print(data[data['Cluster'] == i]['Country'].values)
        print()

    # Visualize the clustering result
    plt.scatter(data['Density'], [0] * len(data), c=data['Cluster'], cmap='viridis')
    plt.yticks([])
    plt.xlabel('Population Density (per sq. mi.)')
    plt.show()

# Load the dataset
data = pd.read_csv('clean_countries.csv', decimal=',')

# Perform clustering and visualization
kmeans_clustering(data, n_clusters=3)
```

Pada syntaks tersebut, fungsi kmeans_clustering digunakan untuk melakukan pengelompokan negara berdasarkan kepadatan penduduk menjadi tiga cluster. 

‚úÖüéÄSetelah melakukan pengelompokan, hasil output ditampilkan pada layar dengan menggunakan print. Dalam hasil output tersebut, terlihat bahwa negara-negara telah berhasil dikelompokkan menjadi tiga kelompok berdasarkan tingkat kepadatan penduduknya.


*   Cluster 0 berisi negara-negara dengan kepadatan penduduk paling tinggi
*   cluster 1 berisi negara-negara dengan kepadatan penduduk sedang

*   cluster 2 berisi negara-negara dengan kepadatan penduduk paling rendah

##Evaluation

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('clean_countries.csv', decimal=',')
```

```python
# Preprocessing data
data = data[['Country', 'Population', 'Area (sq. mi.)']].dropna()
data['Density'] = data['Population'] / data['Area (sq. mi.)']
X = data[['Density']]
```

```python
data = data.dropna() #menghapus baris atau data yang memiliki nilai null 
X = data.drop(['Country'], axis=1) # Drop the 'Country' column
X = (X - X.mean()) / X.std() # melakukan normalisasi atau scaling pada data numerik pada dataset
```

```python
from sklearn.cluster import KMeans

k = 3 #  menentukan jumlah klaster yaitu 3
kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(X) # melakukan fitting atau training model KMeans dengan dataset
```

```python
import matplotlib.pyplot as plt

plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=kmeans.labels_, cmap='viridis')
plt.xlabel(X.columns[0])
plt.ylabel(X.columns[1])
plt.title('K-Means Clustering Result')
plt.show()
```

Dengan visualisasi ini, kita dapat melihat sebaran data pada klaster-klaster yang berbeda dan juga mengevaluasi kualitas klastering dari algoritma K-Means yang digunakan. Jika klaster-klaster tersebut terlihat jelas terpisah dan memiliki karakteristik yang berbeda-beda, maka dapat dikatakan bahwa algoritma K-Means berhasil mengklaster data dengan baik. Namun, jika terdapat banyak titik yang tercampur dan sulit untuk dibedakan, maka dapat dikatakan bahwa klastering tidak terlalu berhasil atau perlu dilakukan perbaikan pada algoritma K-Means yang digunakan.


```python
import pandas as pd
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Calculate silhouette score
silhouette = silhouette_score(X, kmeans.labels_)

# Calculate Calinski-Harabasz index
calinski = calinski_harabasz_score(X, kmeans.labels_)

# Calculate Davies-Bouldin index
davies_bouldin = davies_bouldin_score(X, kmeans.labels_)

# Create a DataFrame to store the scores
scores = pd.DataFrame({'Silhouette Score': [silhouette],
                       'Calinski-Harabasz Score': [calinski],
                       'Davies-Bouldin Score': [davies_bouldin]})

# Display the scores
print(scores)
```

üéÄ‚úÖ‚úÖBerdasarkan output tersebut, kita dapat menarik beberapa kesimpulan sebagai berikut:

1. Silhouette score: Semakin tinggi nilai Silhouette score, semakin baik kualitas cluster yang dihasilkan. Dalam kasus ini, nilai Silhouette score sebesar 0.662 Nilai ini merepresentasikan seberapa dekat suatu titik data dengan klaster yang sama dibandingkan dengan klaster lainnya. Nilai Silhouette Score yang mendekati 1 menunjukkan bahwa pengelompokan atau clustering yang dilakukan sangat baik.
2. Calinski-Harabasz score: Semakin tinggi nilai Calinski-Harabasz score, semakin baik kualitas cluster yang dihasilkan. Dalam kasus ini, nilai Calinski-Harabasz score sebesar 56.102 Nilai ini merepresentasikan seberapa baik atau seberapa terpisah klaster yang dihasilkan oleh model KMeans. Semakin tinggi nilai Calinski-Harabasz Score, semakin terpisah klaster yang dihasilkan.

3. Davies-Bouldin score: Semakin rendah nilai Davies-Bouldin score, semakin baik kualitas cluster yang dihasilkan. Dalam kasus ini, nilai Davies-Bouldin score sebesar 0.636 menunjukkan bahwa kualitas cluster yang dihasilkan cukup baik.
